```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/depositphotos_31129547-stock-photo-venice-landmark-burano-island-canal.jpg',0)
print("Original image:")
cv2_imshow(image)
negative_image = 255 - image
print("Negative Image:")
cv2_imshow(negative_image)
```

```python
lower_threshold = 200
upper_threshold = 255
output_image = np.where((image > lower_threshold) & (image < upper_threshold), 255, 0)
cv2_imshow(output_image)
```

```python
from matplotlib import pyplot as plt
equalized_image = cv2.equalizeHist(image)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.hist(image.ravel(), 256, [0, 256])
plt.title('Original Histogram')
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.subplot(1, 2, 2)
plt.hist(equalized_image.ravel(), 256, [0, 256])
plt.title('Equalized Histogram')
plt.show()
```
