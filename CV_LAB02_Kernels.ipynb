```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread('/content/depositphotos_31129547-stock-photo-venice-landmark-burano-island-canal.jpg')
cv2_imshow(img)
```

Identity kernel

```python
identity=np.array([[1,0,0],[0,1,0],[0,0,1]])
id_img=cv2.filter2D(img,-1,identity)
cv2_imshow(id_img)
```

Edge detection kernel(Scharr)

```python
scharr_img=cv2.Scharr(img,ddepth=-1,dx=1,dy=0)
cv2_imshow(scharr_img)
```

Sharpen kernel

```python
sharp=np.array([[0,-1,0],[-1,6,-1],[0,-1,0]])
sharp_img=cv2.filter2D(img,-1,sharp)
cv2_imshow(sharp_img)
```

BoxFilter kernel

```python
box_img=cv2.boxFilter(img,ddepth=-1,ksize=(10,10))
cv2_imshow(box_img)
```

Gaussian blur kernel

```python
gauss_img=cv2.GaussianBlur(img,ksize=(5,5),sigmaX=30,sigmaY=30)
cv2_imshow(gauss_img)
```
